```python
# Identify potential anomalies
df['price_anomaly'] = df['f_price'] + df['fees'] != df['total_price']

# Filter rows with anomalies (if necessary)
anomalies = df[df['price_anomaly']]
print("Anomalies found:\n", anomalies)
```

```python
df.loc[df['price_anomaly'], 'f_price'] = df['total_price'] - df['fees']
```

```python
# Tương tự cho `f_time_to`
df['arrival_date'] = df['f_time_to'].dt.date
df['arrival_time'] = df['f_time_to'].dt.time
df['arrival_hour'] = df['f_time_to'].dt.hour
df['arrival_day'] = df['f_time_to'].dt.day_name()
df['arrival_month'] = df['f_time_to'].dt.month
df['arrival_year'] = df['f_time_to'].dt.year
```

```python
# Calculate flight time (hours)
df['flight_duration_hours'] = (df['f_time_to'] - df['f_time_from']).dt.total_seconds() / 3600
# Minutes
df['flight_duration_minutes'] = (df['f_time_to'] - df['f_time_from']).dt.total_seconds() / 60
```

```python
# Kiểm tra giá trị không hợp lệ
invalid_dates = df[df['departure_time'] > df['arrival_time']]
print("Các hàng có ngày giờ không hợp lệ:\n", invalid_dates)
```

```python
from geopy.geocoders import Nominatim

# Khởi tạo geolocator
geolocator = Nominatim(user_agent="geoapi")

# Danh sách các tỉnh cần lấy tọa độ
provinces = ["Hà Nội", "TP HCM", "Đà Nẵng", "Hải Phòng", "Thanh Hóa"]

# Hàm lấy tọa độ
def get_coordinates(location):
    try:
        loc = geolocator.geocode(location + ", Vietnam")
        if loc:
            return loc.latitude, loc.longitude
        else:
            return None, None
    except Exception as e:
        print(f"Error for {location}: {e}")
        return None, None

# Lấy tọa độ cho các tỉnh
coordinates = {province: get_coordinates(province) for province in provinces}

# Kết quả
for province, (lat, lon) in coordinates.items():
    print(f"{province}: Latitude = {lat}, Longitude = {lon}")
```
